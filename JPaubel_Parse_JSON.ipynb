```python
import pandas as pd

pd_read_json = pd.read_json('./MOCK_DATA.json')

extract_data = {
    'userNames' : [ uName if uName == 'kstukeg0' else None for uName in pd_read_json['user_name'] ],
    
    'userEmails' : [ uEmail if uEmail else None for uEmail in pd_read_json['user_email'] ],
    
    'bitcoinAddress' : [ bitAdd if bitAdd == '19bQdp5nRv4awMyxbsityVLpUPvtgMCdTb' else None 
                        for bitAdd in pd_read_json['bitcoin_address'] ],
    
    'ipAddress' : [ ipAdd if ipAdd else None for ipAdd in pd_read_json['ip_address'] ],
    
    'timeZone' : [ time if time else None for time in pd_read_json['time_zone'] ],
}

zippedData = list(zip(
    extract_data['userNames'],
    extract_data['userEmails'],
    extract_data['bitcoinAddress'],
    extract_data['ipAddress'],
    extract_data['timeZone'],
))

show_data = pd.DataFrame(zippedData, columns=[
    'User Name', 'UserEmail', 'Bitcoin Address', 'IP Address', 'Time Zone'
]).dropna()

show_data
```
